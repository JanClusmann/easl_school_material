```python
# Online OpenAI LLM Inference
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()  # automatically finds .env in parent dir
api_key = os.getenv("Prevention_OPENAI_Jan")
client = OpenAI(api_key=api_key)

```

```python
#Local
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()  # automatically finds .env in parent dir
api_key = os.getenv("Katherlab_Local")
client = OpenAI(api_key=api_key, base_url="http://pluto/v1/")


```

# Info

```python
print("List of available Models: ")
print(client.models.list())
# Please refer to the LLM model table in Notion! Not all models listed here are currently available!
```

```python
response = client.chat.completions.create(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant!"
        },
        {
            "role": "user",
            "content": "What do you know about the KatherLab?"
        }
    ],
    model="Llama-4-Maverick-17B-128E-Instruct-FP8",
		max_tokens=1024,
        
        
)

print(response.choices[0].message.content)
```
